```python
def gp_tensor():
    dim = 32
    tensor = np.zeros((32,32,32))
    for k in range(dim):
        for i in range(dim):
            for j in range(dim):
                val = gp_table[i,j]
                if abs(val) == k + 1:
                    tensor[k,i,j] = np.sign(val)
    return tensor
def gp_left_matrix(multivector):
    tensor = gp_tensor()
    matrix = np.zeros((32,32))
    for i in range(32):
        t = tensor[i,:,:]
        matrix[i,:] = np.inner(t.T,multivector.T).reshape(-1)
    return matrix
def gp_right_matrix(multivector):
    tensor = gp_tensor()
    matrix = np.zeros((32,32))
    for i in range(8):
        t = tensor[i,:,:]
        matrix[i,:] = np.inner(multivector.T,t).reshape(-1)
    return matrix
```

```python
cga_types = {'sc':0,'e1':1, 'e2':2,'e3':3,'ep':4,'em':5}
ni = np.zeros((32,1))
no = np.zeros((32,1))
# ni = em - ep
# no = 0.5 (em + ep)
ni[cga_types['ep']] = -1.0
ni[cga_types['em']] = 1.0
no[cga_types['ep']] = 0.5
no[cga_types['em']] = 0.5

def make_nullvector(vec):
    nv = np.zeros((32,1))
    nv[cga_types['e1']] = vec[0]
    nv[cga_types['e2']] = vec[1]
    nv[cga_types['e3']] = vec[2]
    nv = nv + 0.5 * (vec[0]*vec[0] + vec[1]*vec[1] + vec[2]*vec[2]) * ni + no
    return nv

vec = [1.,2.,3.]
make_nullvector(vec).reshape(-1)
```
